```python
data_temp = data.copy()
```

```python
data_temp['SeniorCitizen'].apply(lambda val : 'No' if val == 0 else 'Yes')
```

```python
SeniorCitizen_Stream_Yes = data_temp[(data_temp.SeniorCitizen == 1) & (data_temp.StreamingMovies == 'Yes')]['SeniorCitizen'].value_counts() 
NonSeniorCitizen_Stream_Yes = data_temp[(data_temp.SeniorCitizen == 0) & (data_temp.StreamingMovies == 'Yes')]['SeniorCitizen'].value_counts()
```

```python
print('Senior who stream: ', SeniorCitizen_Stream_Yes)
print('Young People who stream: ', NonSeniorCitizen_Stream_Yes)
```

```python
g = pd.DataFrame([SeniorCitizen_Stream_Yes,NonSeniorCitizen_Stream_Yes]).plot(kind="bar", title="Graph showing type of citizens with relation to streaming numbers", legend="True")
g.set_xlabel("Citizens")
g.set_ylabel("Amount Streaming")
g.legend(["No (Non Senior Citizens)","Yes (Senior Citizens)"])
```
